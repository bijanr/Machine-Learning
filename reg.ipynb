```python
import numpy as np  
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
x = np.array([3, 7, 6, 8, 2]).reshape(-1, 1)
y = np.array([5.7, 8, 7, 9, 3]).reshape(-1,1)

plt.scatter(x, y)

```

```python
model = LinearRegression()
model.fit(x, y)
y_pred=model.predict(x)
m = model.coef_[0]
b = model.intercept_
print(f"Intercept: {b}")
print(f"Slope: {m}")

```

```python
# Generate extended x values for a longer regression line
x_extended = np.linspace(0, 6, 100).reshape(-1, 1)  # Extend beyond the original range (0 to 6)

# Compute predictions for the extended x values
y_extended = model.predict(x_extended)

# Visualization
plt.scatter(x, y, color='blue', label='Actual Data')  # Original data points
plt.scatter(0, model.intercept_, color='green', s=100, label="Intercept (1.81)")  # Highlight intercept
plt.plot(x_extended, y_extended, color='red', linestyle='dotted', label='Extended Regression Line')  # Extended regression line
plt.plot(x, y_pred, color='red', linestyle="-", label='Regression Line (Original Range)')  # Original regression line
plt.xlabel('x (Independent Variable)')
plt.ylabel('y (Dependent Variable)')
plt.legend()
plt.title('Extended Linear Regression')
plt.show()
```
